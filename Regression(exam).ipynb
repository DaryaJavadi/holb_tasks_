```python
import pandas as pd
import numpy as np

def remove_outliers_iqr(df, numeric_columns):
    df_cleaned = df.copy()  # copy too change dataframe
    for col in numeric_columns:
        if df_cleaned[col].dtype in ['int64', 'float64']:  # numeric columns
            Q1 = df_cleaned[col].quantile(0.25)
            Q3 = df_cleaned[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
    return df_cleaned

numeric_columns = X.select_dtypes(include=['number']).columns.tolist()

# delete outliers
X = remove_outliers_iqr(X, numeric_columns)
```

#Models

```python
# __Why Did I Use These Models?__

# Linear Regression:
# Reason: It is the simplest and most interpretable model. It is used to model linear relationships and serves as a baseline for small or simple datasets.
# Outcome: Performs well for linear relationships but struggles with non-linear patterns.

# Ridge Regression:
# Reason: It introduces L2 regularization to reduce overfitting, especially in datasets with multicollinearity (high correlation between features).
# Outcome: Improves the model's generalization ability and provides more stable results.

# Lasso Regression:
# Reason: It uses L1 regularization to perform feature selection by shrinking less important feature coefficients to zero.
# Outcome: Simplifies the model and improves interpretability by retaining only the most relevant features.

# Decision Tree Regressor:
# Reason: It models non-linear relationships by splitting the data based on feature values. It is easy to interpret and visualize.
# Outcome: Works well for small to medium-sized datasets but is prone to overfitting.

# Random Forest Regressor:
# Reason: It combines multiple decision trees to reduce overfitting and improve stability. It also provides feature importance scores.
# Outcome: Delivers high accuracy for complex datasets and handles non-linear relationships effectively.

# Polynomial Regression:
# Reason: It extends linear regression by adding polynomial terms to model non-linear relationships.
# Outcome: Effective for small datasets with curved relationships but can overfit if not controlled.
```

```python
# __What Do the Performance Metrics Mean?__

# Mean Absolute Error (MAE):
# Meaning: The average absolute difference between predicted and actual values.
# Purpose: Measures the average magnitude of errors. Lower MAE indicates better performance.

# Mean Squared Error (MSE):
# Meaning: The average squared difference between predicted and actual values.
# Purpose: Penalizes larger errors more heavily. Lower MSE indicates better performance.

# Root Mean Squared Error (RMSE):
# Meaning: The square root of MSE, providing error values in the same units as the target variable.
# Purpose: Easier to interpret than MSE. Lower RMSE indicates better performance.

# R² Score (Coefficient of Determination):
# Meaning: Represents the proportion of variance in the target variable explained by the model.
# Purpose: Ranges from 0 to 1. A value closer to 1 indicates better model performance.
```

```python
# __Why Did We Use These Methods During Data Cleaning, Ensemble, Feature Engineering, and Hyperparameter Tuning?__
# Data Cleaning:
# Methods: Handling missing values (imputation), removing duplicates, encoding categorical variables.
# Why: Ensures the dataset is clean and high-quality, which improves model performance.
# Impact: Increases model accuracy and reduces the risk of overfitting.

# Feature Engineering:
# Methods: Creating new features, transforming existing features, encoding categorical variables.
# Why: Enhances the model's ability to generalize and captures deeper insights from the data.
# Impact: Improves model accuracy and interpretability.

# Ensemble Methods:
# Method: Random Forest.
# Why: Combines multiple models to reduce overfitting and improve stability.
# Impact: Increases generalization ability and delivers higher accuracy.
```

```python
# building a pipeline to handle missing values, scale features, generate polynomial features, and fit a linear regression model
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression

imputer = SimpleImputer(strategy="most_frequent")

poly_pipeline = Pipeline([
    ("imputer", imputer),
    ("scaler", StandardScaler()),
    ("poly_features", PolynomialFeatures(degree=2)),
    ("model", LinearRegression())
])

poly_pipeline.fit(X_train, y_train)
```

```python
poly_pipeline.score(X_train, y_train)
```

```python
# reducing the dimensionality of the dataset to 5 principal components using PCA for both training and testing datasets
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# visualization of the PCA results
import matplotlib.pyplot as plt
import numpy as np

explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(8,5))
plt.bar(range(1, 6), explained_variance * 100, alpha=0.7, color="red", label="Individual Explained Variance")
plt.xlabel("Principal Component")
plt.ylabel("Variance Explained (%)")
plt.title("Explained Variance by PCA Components")
plt.legend()
plt.show()

print(f"Total Explained Variance: {np.sum(explained_variance) * 100:.2f}%")
```

```python
# evaluating the pipeline's performance using cross-validation and calculating the mean R² score
from sklearn.model_selection import cross_val_score

cv_r2_scores = cross_val_score(poly_pipeline, X_train, y_train, cv=5, scoring="r2")

print(f"Cross-validation R² Score: {cv_r2_scores.mean():.4f}")
```

```python
# training a Random Forest Regressor and evaluating its performance using MAE, RMSE, and R² score on the test set
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

ensemble_model = RandomForestRegressor(n_estimators=100, random_state=42)
ensemble_model.fit(X_train, y_train)

y_pred_ensemble = ensemble_model.predict(X_test)
mae_ensemble = mean_absolute_error(y_test, y_pred_ensemble)
rmse_ensemble = mean_squared_error(y_test, y_pred_ensemble)

print(f"Ensemble Model - MAE: {mae_ensemble}, RMSE: {rmse_ensemble}")

r2_ensemble = r2_score(y_test, y_pred_ensemble)
print(f"Ensemble Model: {r2_ensemble:.4f}")
```

```python
# comparing the performance of multiple regression models (Linear, Ridge, Lasso, Decision Tree, Random Forest) using a pipeline with imputation and scaling
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score

models = [
    ('Linear Regression', LinearRegression()),
    ('Ridge Regression', Ridge(alpha=1.0)),
    ('Lasso Regression', Lasso(alpha=0.1)),
    ('Decision Tree Regressor', DecisionTreeRegressor(random_state=42)),
    ('Random Forest Regressor', RandomForestRegressor(n_estimators=100, random_state=42))
]

results = []

for model_name, model in models:
    pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler()),
        ('regressor', model)
    ])

    pipeline.fit(X_train, y_train)

    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)

    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    results.append({
        'Model': model_name,
        'Train MSE': train_mse,
        'Test MSE': test_mse,
        'Test R2 Score': test_r2
    })

results_df = pd.DataFrame(results)

print(results_df)
```
